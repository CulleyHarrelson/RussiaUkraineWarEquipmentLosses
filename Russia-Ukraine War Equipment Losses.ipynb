```python
import pandas as pd

classes_url="https://raw.githubusercontent.com/leedrake5/Russia-Ukraine/main/data/classes.csv"
classes=pd.read_csv(classes_url)
```

```python
totals_by_systems_url="https://raw.githubusercontent.com/scarnecchia/oryx_data/main/totals_by_system.csv"
totals_by_systems=pd.read_csv(totals_by_systems_url)
```

```python
oryx = pd.merge(
    totals_by_systems,
    classes,
    how="inner",
    left_on="system",
    right_on="system"
)
```

```python
#oryx.to_csv('oryx.csv')
oryx.count()
```

```python
import plotly.express as px

#fig = px.strip(oryx, x="date_recorded", y="class", facet_col="country_y")
fig = px.histogram(
    oryx, 
    y="class", 
    facet_col="country_y", 
    color="country_y",
    height=1000,
    labels={"class": "Equipment Category"},
    title="Equipment Losses by Category"
)
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[1]))
fig.show()
```

```python
tanks = oryx[oryx['class'] == 'Tanks']

fig = px.histogram(
    tanks, 
    y="system", 
    facet_col="country_y", 
    height=1100,
    labels={"system": "Type of Tank"},
    title="Tank Losses"
)
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[1]))
fig.show()
```
